```python
oracle1 = oracles.QuadraticOracle(
    np.array([
        [1.0, 2.0], 
        [2.0, 5.0],
    ]), np.zeros(2)
)

plot_levels(oracle1.func)
x_opt, message, history = gradient_descent(oracle1, -np.ones(2) * 3.0, line_search_options = {"method" : "Wolfe"}, trace = True)
plot_trajectory(oracle1.func, history["x"])#, fit_axis=False, label=None)
x_opt, message
```
